```python
import pandas as pd
import numpy as np

data = pd.read_excel("2021 Journal & GL review.xlsx",sheet_name="2.LuxGL")
data
```

```python
data.nunique()
```

```python
columns_to_drop = data.nunique()[data.nunique() == 1].index
columns_to_drop
```

```python
data_dropped_cols = data.copy()
data_dropped_cols = data_dropped_cols.drop(columns=columns_to_drop)
data_dropped_cols
```

```python
data_dropped_cols = data_dropped_cols.dropna(axis=1, how='all')
data_dropped_cols
```

```python
missing_values = [np.nan, "", " ", None]

data_dropped_cols.isin(missing_values).mean().sort_values(ascending=False) * 100
```

```python
columns_to_drop = data_dropped_cols.isin(missing_values).mean()[(data_dropped_cols.isin(missing_values).mean()) > 0.86].index
data_missing_values = data_dropped_cols.copy()
data_missing_values = data_dropped_cols.drop(columns=columns_to_drop)
data_missing_values
```
