```python
import pandas as pd

data = pd.read_excel("ML Test BBW.xlsx")
data
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder()
columns_to_encode = ['DIM4', 'DIM7', 'EnteredBy', 'Username']
onehot_encoder.fit(data[columns_to_encode])
encoded_data = onehot_encoder.transform(data[columns_to_encode]).toarray()

new_columns = onehot_encoder.get_feature_names_out(columns_to_encode)
encoded_df = pd.DataFrame(encoded_data, columns=new_columns)
print(encoded_df.head())
```

```python
# Drop original columns
data_dropped = data.drop(columns=columns_to_encode)

# Concatenate the original DataFrame and the one-hot encoded DataFrame
final_data = pd.concat([data_dropped, encoded_df], axis=1)
final_data
```

```python
actual_columns = final_data.columns.tolist()
actual_columns
```

```python
from sklearn.model_selection import train_test_split

x = final_data[["VatRate1", "DIM4_BBW-AUOL  ", "DIM4_BBW-CSM   ",
"DIM4_BBW-DFO   ", "DIM4_BBW-QVB   ", "DIM4_BBW-WCS   ",
"DIM4_BBW-WP    ", "DIM7_OP20-101  ", "DIM7_OP20-120  ",
"DIM7_OP20-131  ", "DIM7_OP20-132  ", "DIM7_OP20-133  ",
"EnteredBy_MOVEX     ", "EnteredBy_SV.M3API  ", "EnteredBy_SV.MECAPI ",
"Username_M3 API for Apps                         ",
"Username_M3 Auto Start Job                       ",
"Username_MEC API                                 "]]
y = final_data["LocalAmt"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=87)
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
y_pred = lr_model.predict(x_test)

pd.DataFrame({
    "y_pred": y_pred,
    "y_test": y_test
})
```
